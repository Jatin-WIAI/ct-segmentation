```python
import numpy as np
import nibabel as nib

from glob import glob

import matplotlib.pyplot as plt
```

```python
ROOT_DIR = '/scratche/users/sansiddh/abdomenCT-1k/'
all_images = glob(ROOT_DIR+'Cases/*.nii.gz')
all_masks = glob(ROOT_DIR+'Masks/*.nii.gz')
len(all_images), len(all_masks)
```

```python
ctscan = nib.load(all_images[0]).get_fdata()
mask = nib.load(all_masks[0]).get_fdata()
```

```python
fig, axs = plt.subplots(figsize=(25, 25), nrows=5, ncols=5)
for i in range(25):
    ax = axs.flat[i]
    ax.imshow(np.rot90(ctscan[:, :, i*10], k=1), cmap='gray')
```

```python
fig, axs = plt.subplots(figsize=(25, 25), nrows=5, ncols=5)
for i in range(25):
    ax = axs.flat[i]
    ax.imshow(np.rot90(mask[:, :, i*10], k=1), cmap='gray')
    print(np.unique(mask[:, :, i*10]))
```
